```python
from keras.applications import ResNet50
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Flatten , Dense , Dropout,BatchNormalization
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import glob
import cv2
import os
import warnings
warnings.simplefilter('ignore')
```

```python
image_height,image_width=200,200
number_class=11
```

```python
def preProcess(img, sigma):   
    height, width, depth = img.shape  
    x = int(width/2)
    y = int(height/2)  
    circle_img = np.zeros((height, width), np.uint8)
    cv2.circle(circle_img, (x,y), min(x,y), 1, thickness=-1)
    img = cv2.bitwise_and(img, img, mask=circle_img) 
    grayImage = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) 
    x, y = np.nonzero(grayImage)  
    img=img[x.min():x.max(), y.min():y.max(),:]
    if img.shape[0]<1000:
        img = cv2.resize(img, (2000,2000))
    img=cv2.addWeighted(img,4,cv2.GaussianBlur( img , (0,0),sigma) ,-4 ,30)
    return img 
```

```python
def get_images(disease,path):
    data = glob.glob(path+'\\%s\\*' % disease)
    return data
```

```python
def generate_dataset(disease,path):         
    for dis in disease:
        data = get_images(dis,path)
        i=0  
        for img in data:
            img =cv2.imread(img)      
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = preProcess(img,50)
            img = cv2.resize(img, (image_height,image_width))    
            img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
            if not os.path.exists('dataGene\\'+dis):
                os.mkdir('dataGene\\'+dis)
            cv2.imwrite('dataGene\\'+dis+'\\'+str(disease.index(dis))+str(i)+".jpg",img ) 
            i=i+1
```

```python
def get_dataset(disease,path):
    
    dataGen = []
    labels = []           
    for dis in disease:
        data = get_images(dis,path)
        for img in data:
            img=cv2.imread(img)      
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img= cv2.resize(img, (image_height,image_width)) 
            dataGen.append(img)
            labels.append(disease.index(dis))    

    return dataGen, labels
```

```python
diseases=['Chorioretinal atrophy-coloboma','CRVO','DR3','Massive hard exudates','Normal','Optic atrophy','Pathological myopia','Retinitis pigmentosa','Tessellated fundus','Vessel tortuosity','Severe hypertensive retinopathy']
```

```python
X,y = get_dataset(diseases,'dataGene')
```

```python
y=np.array(y)
X=np.array(X)
```

```python
import seaborn as sns
sns.countplot(y)
```

```python
y=np.array(y)
y=y.reshape(y.shape[0],-1)
X=np.array(X)
print('y=',y.shape)
print('x=',X.shape)
```

```python
X_train_orig,X_test_orig, y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
print(y_train.shape)
print(X_test_orig.shape)
```

```python
model=ResNet50(weights='imagenet',
               include_top=False,
               input_shape=(image_height,image_width,3),

)
```

```python
model.save("model_resnet50.h5")
```

```python
fe_train=model.predict(X_train_orig)
```

```python
fe_test=model.predict(X_test_orig)
```

```python
fe_train.shape 
```

```python
model_2=Sequential()
model_2.add(Flatten(input_shape=fe_train.shape[1:]))
model_2.add(Dense(100,activation='relu'))
model_2.add(BatchNormalization())
model_2.add(Dense(50,activation='relu'))
model_2.add(BatchNormalization())
model_2.add(Dense(11,activation='softmax'))
```

```python
model_2.compile('adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
```

```python
model_2.summary()
```

```python
his=model_2.fit(fe_train,y_train,epochs=10,batch_size=150,validation_split=0.25)
```

```python
plt.plot(his.history['acc'])
plt.plot(his.history['val_acc'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('model accuracy')
plt.legend(['train','validation'])
plt.show()
```

```python
score=model_2.evaluate(fe_test,y_test)
print('acc',score[1])
```

```python
model_2.save("model.h5")
```
